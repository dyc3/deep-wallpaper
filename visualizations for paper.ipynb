# Visualizations

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Decision tree visuals

```python
reds_x = np.concatenate((np.random.uniform(0, 12, (100, 1)), np.random.uniform(8, 20, (100, 1))))
reds_y = np.concatenate((np.random.uniform(0, 12, (100, 1)), np.random.uniform(14.5, 20, (100, 1))))
greens_x = np.concatenate((np.random.uniform(13.5, 20, (80, 1)), np.random.uniform(0, 6, (50, 1))))
greens_y = np.concatenate((np.random.uniform(0, 11.5, (80, 1)), np.random.uniform(13.5, 20, (50, 1))))
```

```python
plt.scatter(reds_x, reds_y, c='r')
plt.scatter(greens_x, greens_y, c='g')

plt.plot([-5, 25], [12.5, 12.5], label="Split 1", c="r", linestyle="--")
plt.plot([12.5, 12.5], [-5, 12.5], label="Split 2", c="g", linestyle="--")
plt.plot([6.5, 6.5], [12.5, 25], label="Split 3", c="b", linestyle="--")
plt.axis([0,20,0,20])
plt.annotate("Split 1", xy=(17, 13), size="x-large", color="r")
plt.annotate("Split 2", xy=(12.7, 9), size="x-large", color="g", rotation="vertical")
plt.annotate("Split 3", xy=(6.75, 17), size="x-large", color="b", rotation="vertical")
```
